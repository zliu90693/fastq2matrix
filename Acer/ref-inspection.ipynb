↑ There is as many rows of non-empty `gene_biotype` as there are `gene_ids`. In other words, in each of the rows corresponding to a `gene_id`, there is one row of non-empty `gene_biotype`.

↓ within each `gene_id` group where only one row has a non-empty `gene_biotype` and the rest are empty strings, fill all rows of that group with that single non-empty value.

My approach was to manually parse and modify the GTF file. The original method involved using gtfparse to read the file, then filling in the `gene_biotype` field, and finally writing it as a GTF file using pyranges (because pyranges cannot read it directly). However, I later discovered this approach didn't work because gtfparse would parse the sixth column (`score`) of the GTF file as null (in GTF, its content is `.`), and it would also parse all the `.` in the eighth column (`frame`) as 0. As a result, when pyranges wrote the file as GTF, `cellranger mkref` couldn't parse it at all.

```python
import re
from collections import defaultdict
```

```python
INPUT_GTF  = "./ref/GCF_029169275.1_AcerK_1.0_genomic_corrected.gtf"
OUTPUT_GTF = "./ref/GCF_029169275.1_fixed.gtf"
```

```python
# ── 第一步：解析GTF，原样保留每一行 ──────────────────────────────────────────

def parse_attributes(attr_str):
    """将attribute字符串解析为有序列表[(key, value), ...]，完整保留原始内容"""
    result = []
    for kv in attr_str.rstrip("; ").split("; "):
        kv = kv.strip()
        if not kv:
            continue
        parts = kv.split(" ", 1)
        if len(parts) == 2:
            key = parts[0]
            val = parts[1].strip('"')
            result.append((key, val))
    return result

def attributes_to_str(attr_list):
    """将[(key, value), ...]还原为GTF第9列字符串"""
    return "; ".join(f'{k} "{v}"' for k, v in attr_list) + ";"
```

```python
# 读取所有行，保留注释行，解析数据行
header_lines = []
data_rows = []   # 每个元素: {"fields": [...8个字段...], "attrs": [(k,v),...]}

with open(INPUT_GTF, "r") as f:
    for line in f:
        line = line.rstrip("\n")
        if line.startswith("#"):
            header_lines.append(line)
            continue
        fields = line.split("\t")
        if len(fields) < 9:
            continue
        attrs = parse_attributes(fields[8])
        data_rows.append({
            "fields": fields[:8],   # 原样保留前8列，含"."不做任何转换
            "attrs":  attrs
        })
```

```python
# ── 第二步：收集每个 gene_id 对应的 gene_biotype ──────────────────────────────

gene_biotype_map = {}   # gene_id → gene_biotype

for row in data_rows:
    feature = row["fields"][2]
    if feature != "gene":
        continue
    attr_dict = dict(row["attrs"])
    gid = attr_dict.get("gene_id", "")
    gbt = attr_dict.get("gene_biotype", "")
    if gid and gbt:
        gene_biotype_map[gid] = gbt

print(f"A total of {len(gene_biotype_map)} gene_ids and their corresponding gene_biotypes were collected.")
```

```python
# ── 第三步：填充 gene_biotype，并写出 GTF ─────────────────────────────────────

filled_count = 0

with open(OUTPUT_GTF, "w") as out:
    # 写文件头
    for h in header_lines:
        out.write(h + "\n")

    for row in data_rows:
        fields  = row["fields"]
        attrs   = row["attrs"]     # [(key, val), ...]
        feature = fields[2]

        attr_dict = dict(attrs)
        gene_id   = attr_dict.get("gene_id", "")

        # 非gene行，且缺少gene_biotype，且能查到对应值 → 填充
        if (feature != "gene"
                and "gene_biotype" not in attr_dict
                and gene_id in gene_biotype_map):

            # 插入位置：紧跟在 gene_id 之后（或追加到末尾）
            new_attrs = []
            inserted  = False
            for k, v in attrs:
                new_attrs.append((k, v))
                if k == "gene_id" and not inserted:
                    new_attrs.append(("gene_biotype", gene_biotype_map[gene_id]))
                    inserted = True
            if not inserted:
                new_attrs.append(("gene_biotype", gene_biotype_map[gene_id]))
            attrs = new_attrs
            filled_count += 1

        attr_str = attributes_to_str(attrs)
        out.write("\t".join(fields) + "\t" + attr_str + "\n")

print(f"A total of {filled_count} rows of gene_biotype were populated.")
```
