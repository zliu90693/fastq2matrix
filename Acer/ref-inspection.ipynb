To ensure the generality of the process (i.e., [2_make-ref.sh](../2_make-ref.sh) filters for `gene_biotype` rather than `transcript_biotype`), I need to fill in the missing `gene_biotype` values ​​in each line of the current GTF file.

```python
import pyranges as pr
import pandas as pd
from gtfparse import read_gtf
```

```python
Acer_df = read_gtf("./ref/GCF_029169275.1_AcerK_1.0_genomic_corrected.gtf") # 直接用pyranges读取会出错, 原因未知
Acer_df
```

```python
Acer_df = Acer_df.to_pandas()
```

```python
print((Acer_df["gene_biotype"] != "").sum())
print(Acer_df["gene_id"].nunique())

# print(Acer_df["gene_biotype"].unique())
```

↑ There is as many rows of non-empty `gene_biotype` as there are `gene_ids`. In other words, in each of the rows corresponding to a `gene_id`, there is one row of non-empty `gene_biotype`.

↓ For a pandas DataFrame, within each `gene_id` group where only one row has a non-empty `gene_biotype` and the rest are empty strings, fill all rows of that group with that single non-empty value.

```python
Acer_df['gene_biotype'] = Acer_df['gene_biotype'].replace('', pd.NA)
#    transform('first') 会跳过 NaN 取第一个有效值，若全为空则返回 NaN: 
Acer_df['gene_biotype'] = Acer_df.groupby('gene_id')['gene_biotype'].transform('first')
```

```python
print(Acer_df["gene_biotype"].isna().sum())
```

↑ The replacement has been completed.

```python
"gene_biotype" in Acer_df.columns
```

```python
Acer_df = Acer_df.rename(columns={"seqname": "Chromosome", # 必须重命名, 否则无法转换
                        "start": "Start",
                        "end": "End"})
Acer_df["Start"] = Acer_df["Start"] - 1 # 后续转换成 PyRanges 再写成 GTF 会导致 Start 全体加一, 所以必须在这里预先减一
# new_pr = pr.PyRanges(Acer_df)
```
