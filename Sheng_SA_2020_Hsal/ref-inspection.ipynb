# 2

Examine the CDS and exon start site:

```python
%%bash

gppy txinfo -g "./ref/Harpegnathos_saltator.gtf" > ./metadata/Hsal_txinfo.tsv
```

```python
import pandas as pd

Hsal_txinfo = pd.read_csv("./metadata/Hsal_txinfo.tsv", sep="\t")
Hsal_txinfo
```

```python
utr5_missing = ((Hsal_txinfo["utr5_len"] == 0) & 
                (Hsal_txinfo["transcript_biotype"] == 'protein_coding')).sum() / \
               (Hsal_txinfo["transcript_biotype"] == 'protein_coding').sum()
print(utr5_missing)
```

```python
utr3_missing = ((Hsal_txinfo["utr3_len"] == 0) & 
                (Hsal_txinfo["transcript_biotype"] == 'protein_coding')).sum() / \
               (Hsal_txinfo["transcript_biotype"] == 'protein_coding').sum()
print(utr3_missing)
```

```python
both_missing = ((Hsal_txinfo["utr3_len"] == 0) & 
                (Hsal_txinfo["utr5_len"] == 0) &
                (Hsal_txinfo["transcript_biotype"] == 'protein_coding')).sum() / \
               (Hsal_txinfo["transcript_biotype"] == 'protein_coding').sum()
print(both_missing)
```

```python
either_missing = (((Hsal_txinfo["utr3_len"] == 0) | (Hsal_txinfo["utr5_len"] == 0)) & # 注意!! 必须将 | 括起来, 因为 & 的优先级比 | 大
                (Hsal_txinfo["transcript_biotype"] == 'protein_coding')).sum() / \
               (Hsal_txinfo["transcript_biotype"] == 'protein_coding').sum()
print(either_missing)
```
