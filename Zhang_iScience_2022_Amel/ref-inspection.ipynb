# 2

Since the analysis of non-type species heavily relies on the annotation of UTR sequences in the GTF, the CDS and exon start sites are compared here. If they are the same, it means that the exon annotation does not contain the UTR region and cannot be used directly; if they are different, it means that the exon annotation contains the UTR region. This is verified using gppy.

```python
%%bash

gppy txinfo -g ./ref/Apis_mellifera.Amel_HAv3.1.62.gtf > ./metadata/Amel_txinfo.tsv
```

```python
import pandas as pd
```

```python
Amel_txinfo = pd.read_csv("./metadata/Amel_txinfo.tsv", sep="\t")
Amel_txinfo
```

```python
Amel_txinfo["gene_biotype"].unique()
```

```python
# ((Amel_txinfo["utr3_len"] == 0) & (Amel_txinfo["gene_biotype"] == 'protein_coding')).sum() / (Amel_txinfo["gene_biotype"] == 'protein_coding').sum()
# 一个gene可能包含多个transcript, 而UTR与transcript而非gene一一对应, 建议用transcript而非gene ↓
```

```python
utr5_missing = ((Amel_txinfo["utr5_len"] == 0) & 
                (Amel_txinfo["gene_biotype"] == 'protein_coding')).sum() / \
               (Amel_txinfo["gene_biotype"] == 'protein_coding').sum()
print(utr5_missing)
```

```python
utr3_missing = ((Amel_txinfo["utr3_len"] == 0) & 
                (Amel_txinfo["gene_biotype"] == 'protein_coding')).sum() / \
               (Amel_txinfo["gene_biotype"] == 'protein_coding').sum()
print(utr3_missing)
```

```python
both_missing = ((Amel_txinfo["utr3_len"] == 0) & 
                (Amel_txinfo["utr5_len"] == 0) &
                (Amel_txinfo["gene_biotype"] == 'protein_coding')).sum() / \
               (Amel_txinfo["gene_biotype"] == 'protein_coding').sum()
print(both_missing)
```
